```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open('DOGEx_v1/ajgj.neu') as to_read:
    lines = to_read.readlines()
```

```python
with open("DOGEx_v1/csv/ajgj.csv", "a") as to_append:
    for line in lines:
        line_reconstructed = ''

        items = line.split(' ')
        items = [item for item in items if item != '']

        for item in items:
            line_reconstructed += item + ',' 
        line_reconstructed = line_reconstructed[0:-1]

        to_append.write(line_reconstructed)
```

```python
df = pd.read_csv('DOGEx_v1/csv/awmi.csv',sep=',', header=None)
df.columns = ['date', 'un', 'ue', 'uz', 'sn', 'se', 'sz']
df.set_index('date', inplace=True)
df.drop(['sn', 'se', 'sz'], axis=1, inplace=True)
```

### Graphing deltas 

```python
plt.plot(df, label=['un', 'ue', 'uz']);
plt.xticks(rotation=25)
plt.legend();
```
